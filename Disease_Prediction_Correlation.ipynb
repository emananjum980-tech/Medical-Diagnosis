```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
print("üìÅ Please upload your cleaned dataset (heart_cleaned.csv)...")
uploaded = files.upload()
df = pd.read_csv("heart_cleaned.csv")
print("‚úÖ Dataset Loaded Successfully!")
print("Shape:", df.shape)
print("Columns:", list(df.columns), "\n")
print("üìà Basic Statistics:\n")
for col in df.select_dtypes(include=np.number).columns:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode()[0]
    variance_val = df[col].var()
    print(f"{col}:")
    print(f"   Mean     = {mean_val:.2f}")
    print(f"   Median   = {median_val}")
    print(f"   Mode     = {mode_val}")
    print(f"   Variance = {variance_val:.2f}\n")
print("üîó Correlation Matrix with Target Variable:\n")
corr_matrix = df.corr()
print(corr_matrix['target'].sort_values(ascending=False))
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Features")
plt.show()
target_corr = corr_matrix['target'].abs().sort_values(ascending=False)
top_features = target_corr[1:4]  # Exclude target itself
print("\nüíé Top 3 Features Most Related to Target Variable:")
print(top_features, "\n")
print("üìÑ Summary Report:")
print("Based on correlation analysis, the three features most strongly associated "
      "with heart disease (target variable) are:\n")
for feature, corr_value in top_features.items():
    print(f"‚Ä¢ {feature}: correlation = {corr_value:.2f}")
print("\nThese features are likely the key predictors for heart disease and "
      "should be prioritized in model training.")
```
