```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
print("üìÅ Please upload your cleaned dataset file (heart_cleaned.csv)...")
uploaded = files.upload()
df = pd.read_csv("heart_cleaned.csv")
print("‚úÖ Dataset Loaded Successfully!")
print(df.head(), "\n")
print("Shape:", df.shape)
print("Columns:", list(df.columns))
sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (8, 5)
plt.title("Heart Disease Distribution (0 = No, 1 = Yes)")
sns.countplot(x="target", data=df)
plt.xlabel("Heart Disease Presence")
plt.ylabel("Number of Patients")
plt.show()
print("üí° Insight 1:")
print("Most patients in the dataset have heart disease (target = 1). "
      "This indicates a slight imbalance, but also highlights prevalence in data.\n")
plt.title("Age Distribution of Patients")
sns.histplot(df["age"], kde=True, bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("üí° Insight 2:")
print("Most patients are between 40 and 60 years old ‚Äî this age group is at higher risk of heart disease.\n")
plt.title("Age vs Cholesterol Levels")
sns.scatterplot(x="age", y="chol", hue="target", data=df, alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
print("üí° Insight 3:")
print("Higher cholesterol levels are visible in older patients, "
      "and people with heart disease (target=1) often have higher cholesterol.\n")
plt.title("Chest Pain Type vs Heart Disease")
sns.countplot(x="cp", hue="target", data=df)
plt.xlabel("Chest Pain Type (cp)")
plt.ylabel("Count")
plt.legend(["No Disease", "Disease"])
plt.show()
print("üí° Insight 4:")
print("Certain chest pain types (like cp=2 or cp=3) are more associated with heart disease cases.\n")
plt.title("Correlation Heatmap of Numeric Features")
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
print("üí° Insight 5:")
print("Strong positive correlation is observed between 'cp', 'thalach', and 'target', "
      "while 'oldpeak' shows negative correlation ‚Äî important for model prediction.\n")
```
